1. Escribe un programa que pida al usuario un texto (como si fuera un tweet o un post de Instagram) y extraiga automáticamente solo las palabras que son hashtags (es decir, las que empiezan con el símbolo #).

```python
# Manera clasica 
entrada = input("Agrega tu comentario")
 
entrada = entrada.split()
lista = []
for tupla in entrada:
    if tupla[0] == "#":
        lista.append(tupla)
print(lista)
```

```python
# Manera de lambda

entrada = input("Agrega tu comentario")
 
entrada = entrada.split()

lista = list(filter(lambda tupla: tupla[0]=="#", entrada))
```

```python
# Manera: list comprehesion

entrada = input("Agrega tu comentario")
 
entrada = entrada.split()

lista = [palabra for palabra in entrada if palabra[0]== "#"]

print(lista)
```

2. Extraer hashtags, pero con dos condiciones:

- Debe empezar con # (como antes).

- Debe tener más de 3 caracteres de longitud (ej. #py no cuenta, pero #python sí).

```python
# Opcion 1

entrada = input("Agrega tu comentario")
 
entrada = entrada.split()
lista = []
for palabra in entrada:
    if palabra[0] == "#" and len(palabra) > 3:
        lista.append(palabra)
print(lista)
```

```python
# Opcion 2
entrada = input("Agrega tu comentario")
entrada = entrada.split()
 
lista = list(filter(lambda palabra: palabra[0] == "#" and len(palabra)>3, entrada))

print(lista)
```

```python
# Opcion 3
entrada = input("Agrega tu comentario")
 
entrada = entrada.split()

lista = [palabra for palabra in entrada if palabra[0] == "#" and len(palabra) > 3]
print(lista)
```

3.Crear una nueva lista llamada inventario_procesado que contenga solo los productos que cuestan más de 1.20.

Pero el formato de la nueva lista debe ser una lista de diccionarios, de la siguiente manera:

```python

[
    {'id': 'P01', 'producto': 'Manzana', 'precio': 1.50},
    {'id': 'P02', 'producto': 'Leche', 'precio': 2.50},
    {'id': 'P04', 'producto': 'Pollo', 'precio': 7.80}
]
```

```python
# Lista de listas: [ID_Producto, Nombre, Categoria, Precio]
inventario_lista = [
    ["P01", "Manzana", "Fruta", 1.50],
    ["P02", "Leche", "Lácteo", 2.50],
    ["P03", "Pan", "Panadería", 1.00],
    ["P04", "Pollo", "Carne", 7.80],
    ["P05", "Lechuga", "Verdura", 0.90]
]
```

```python
inventario_procesado = list(filter(lambda producto: producto[3] > 1.2, inventario_lista))
print(inventario_procesado)

inventario_procesado2 =[]
numeros_exactos = [0,1,3]
for lista in inventario_procesado:
    inventario_p = list(lista[i] for i in numeros_exactos)
    inventario_procesado2.append(inventario_p)
print(inventario_procesado2)

categorias = ["ID_Producto", "Nombre", "Precio"]
print(categorias)


lista_diccionario = []
for lista in inventario_procesado2:
    diccionario = {categorias[i]: lista[i] for i in range(len(categorias))}
    lista_diccionario.append(diccionario)

print(lista_diccionario)

```

```python
import json  # Importamos la librería para manejar archivos JSON

# ... (Aquí va todo tu código anterior que genera lista_diccionario) ...

# Nombre del archivo donde guardaremos los datos
nombre_archivo = "mi_inventario_filtrado.json"

# Abrimos (o creamos) el archivo en modo escritura ('w' de write)
with open(nombre_archivo, 'w') as archivo:
    # json.dump volca tus datos al archivo
    # indent=4 sirve para que se vea bonito y ordenado con sangría
    json.dump(lista_diccionario, archivo, indent=4)

print(f"¡Listo! Datos guardados en {nombre_archivo}")
```
